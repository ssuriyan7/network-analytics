```python
class_df = all_df['class']
class_df = pd.factorize(class_df)[0]
```

```python
interface_df = pd.concat([interface_df, class_df['class']],axis=1)
tcp_df = pd.concat([tcp_df, class_df['class']],axis=1)
udp_df = pd.concat([udp_df, class_df['class']],axis=1)
ip_df = pd.concat([ip_df, class_df['class']],axis=1)
icmp_df = pd.concat([icmp_df, class_df['class']],axis=1)
```

```python
#ip_df.head()
#interface_df.head()
#tcp_df.head()
#udp_df.head()
#icmp_df.head()
```

```python
#ip category
#input features

X_interface = interface_df[['ifInOctets11','ifOutOctets11','ifoutDiscards11','ifInUcastPkts11','ifInNUcastPkts11','ifInDiscards11','ifOutUcastPkts11','ifOutNUcastPkts11']]
X_tcp = tcp_df[['tcpOutRsts','tcpInSegs','tcpOutSegs','tcpPassiveOpens','tcpRetransSegs','tcpCurrEstab','tcpEstabResets','tcp?ActiveOpens']]
X_udp = udp_df[['udpInDatagrams','udpOutDatagrams','udpInErrors','udpNoPorts']]
X_ip = ip_df[['ipInReceives','ipInDelivers','ipOutRequests','ipOutDiscards','ipInDiscards','ipForwDatagrams','ipOutNoRoutes','ipInAddrErrors']]
X_icmp = icmp_df[['icmpInMsgs','icmpInDestUnreachs','icmpOutMsgs','icmpOutDestUnreachs','icmpInEchos','icmpOutEchoReps']]


#output class labels
y = ip_df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_interface_train, X_interface_test, y_interface_train, y_interface_test = train_test_split(X_interface, y, test_size=0.2)
X_tcp_train, X_tcp_test, y_tcp_train, y_tcp_test = train_test_split(X_tcp, y, test_size=0.2)
X_udp_train, X_udp_test, y_udp_train, y_udp_test = train_test_split(X_udp, y, test_size=0.2)
X_ip_train, X_ip_test, y_ip_train, y_ip_test = train_test_split(X_ip, y, test_size=0.2)
X_icmp_train, X_icmp_test, y_icmp_train, y_icmp_test = train_test_split(X_icmp, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_interface=RandomForestClassifier(n_estimators=100)
clf_tcp=RandomForestClassifier(n_estimators=100)
clf_udp=RandomForestClassifier(n_estimators=100)
clf_ip=RandomForestClassifier(n_estimators=100)
clf_icmp=RandomForestClassifier(n_estimators=100)
```

```python
clf_interface.fit(X_interface_train,y_interface_train)
```

```python
clf_tcp.fit(X_tcp_train,y_tcp_train)
```

```python
clf_udp.fit(X_udp_train,y_udp_train)
```

```python
clf_ip.fit(X_ip_train,y_ip_train)
```

```python
clf_icmp.fit(X_icmp_train,y_icmp_train)
```

```python
y_interface_pred=clf_interface.predict(X_interface_test)
y_tcp_pred=clf_tcp.predict(X_tcp_test)
y_udp_pred=clf_udp.predict(X_udp_test)
y_ip_pred=clf_ip.predict(X_ip_test)
y_icmp_pred=clf_icmp.predict(X_icmp_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy for :",metrics.accuracy_score(y_interface_test, y_interface_pred))
print("Accuracy:",metrics.accuracy_score(y_tcp_test, y_tcp_pred))
print("Accuracy:",metrics.accuracy_score(y_udp_test, y_udp_pred))
print("Accuracy:",metrics.accuracy_score(y_ip_test, y_ip_pred))
print("Accuracy:",metrics.accuracy_score(y_icmp_test, y_icmp_pred))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
from matplotlib.colors import ListedColormap


def plot_classification_report(classificationReport,
                               title='Classification report',
                               cmap='RdBu'):
    
    colors = ['salmon']
    cmap = ListedColormap(colors)
    classificationReport = classificationReport.replace('\n\n', '\n')
    classificationReport = classificationReport.replace(' / ', '/')
    lines = classificationReport.split('\n')

    classes, plotMat, support, class_names = [], [], [], []
    for line in lines[1:]:  # if you don't want avg/total result, then change [1:] into [1:-1]
        t = line.strip().split()
        if len(t) < 2:
            continue
        classes.append(t[0])
        v = [float(x) for x in t[1: len(t) - 1]]
        support.append(int(t[-1]))
        class_names.append(t[0])
        plotMat.append(v)

    plotMat = np.array(plotMat)
    xticklabels = ['Precision', 'Recall', 'F1-score']
    yticklabels = ['{0} ({1})'.format(class_names[idx], sup)
                   for idx, sup in enumerate(support)]

    plt.imshow(plotMat, interpolation='nearest', cmap=cmap, aspect='auto')
    plt.title(title)
    #plt.colorbar()
    plt.xticks(np.arange(3), xticklabels, rotation=45)
    plt.yticks(np.arange(len(classes)), yticklabels)

    upper_thresh = plotMat.min() + (plotMat.max() - plotMat.min()) / 10 * 8
    lower_thresh = plotMat.min() + (plotMat.max() - plotMat.min()) / 10 * 2
    for i, j in itertools.product(range(plotMat.shape[0]), range(plotMat.shape[1])):
        plt.text(j, i, format(plotMat[i, j], '.2f'),
                 horizontalalignment="center",
                 color="black")

    plt.ylabel('Classes')
    plt.xlabel('Metrics')
    plt.tight_layout()
```

```python
plot_classification_report(classification_report(y_interface_test, y_interface_pred))
plt.show()
```

```python
plot_classification_report(classification_report(y_tcp_test, y_tcp_pred))
plt.show()
```
